```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
import pickle
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction import DictVectorizer
```

## Read the data

```python
taxi01 = pd.read_parquet('data/yellow_tripdata_2022-01.parquet')
```

## Q1

```python
taxi01.shape
```

## Q2

```python
taxi01.info()
```

```python
taxi01['duration'] = (taxi01['tpep_dropoff_datetime'] - taxi01['tpep_pickup_datetime'])

taxi01['duration']  = taxi01['duration'].apply(lambda x: x.total_seconds() / 60)

taxi01['duration'].std()
```

## Q3

```python
taxi01_update = taxi01.loc[(taxi01['duration'] >= 1) & (taxi01['duration'] <= 60)]
```

```python
taxi01_update.shape[0] / taxi01.shape[0]
```

## Q4

```python
taxi01_update.columns
```

```python
taxi01_selected = taxi01_update[[ 'PULocationID', 'DOLocationID']]
taxi01_selected[[ 'PULocationID', 'DOLocationID']] = taxi01_selected[[ 'PULocationID', 'DOLocationID']].astype(str)
```

```python
taxi01_selected.info()
```

```python
ohe = OneHotEncoder()

taxi01_selected = ohe.fit_transform(taxi01_selected)
```

```python
dvt = DictVectorizer()

taxi01_dict = taxi01_selected.to_dict(orient='records')

taxi01_dict
```

```python
train_data = dvt.fit_transform(taxi01_dict)
train_data.shape
```

## Q5

```python
train_y = taxi01_update['duration']

lr = LinearRegression()

lr.fit(train_data, train_y)

print(f'RMSE of train data: {mean_squared_error(train_y, lr.predict(train_data))**0.5}')
```
